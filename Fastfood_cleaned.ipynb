```python
# importing the dependencies
import pandas as pd
```

```python
# importing the data into Pandas
df = pd.read_csv("./Resources/Fast_Food_Restaurants_US.csv")
df
```

```python
# dropping the colums not required for the cluster map
df_cleaned= df.drop(columns=['Unnamed: 0','categories','city','country','postalCode','websites'])
df_cleaned
```

```python
#Nomarlize the names of the restaurants
df_cleaned['name'] = df_cleaned['name'].str.replace(".", "", regex=True)
df_cleaned['name'] = df_cleaned['name'].str.replace("'","", regex=True)
df_cleaned['name'] = df_cleaned['name'].str.replace(" ","", regex=True)
df_cleaned['name'] = df_cleaned['name'].str.replace(".","", regex=True)
df_cleaned['name'] = df_cleaned['name'].str.replace("-","", regex=True)
df_cleaned['name'] = df_cleaned['name'].str.replace("&","AND", regex=True)
df_cleaned['name'] = df_cleaned['name'].str.upper()
```

```python
restaurants_count_df = df_cleaned['province'].value_counts().rename_axis('State').reset_index(name='counts')
restaurants_count_df.sort_values(by='State').set_index('State').head()
```

```python
restaurants_count_df
```

```python
restaurants_count_df = restaurants_count_df.rename(columns={'State': 'state_code'})
restaurants_count_df.head()
```

```python
restaurants_count_df.to_csv("clean_data/Fast_Food_Restaurants_US_cleaned.csv", index=False, header=True)
```
