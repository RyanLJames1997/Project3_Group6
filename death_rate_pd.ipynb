```python
from pathlib import Path
import pandas as pd
```

```python
# Create a reference to the file.
csv_path = Path("Resources/number-of-deaths-by-risk-factor (2).csv")
# Read the data into a DataFrame. 
death_df = pd.read_csv(csv_path, encoding="UTF-8")
death_df.head() 
```

```python
death_df["Entity"].unique()
```

```python
grouped_entity = death_df.groupby('Entity')
grouped_entity.head()
```

```python
usa_grouped = grouped_entity.get_group('United States')
usa_grouped
```

```python
usa_grouped = usa_grouped.tail(6)
usa_grouped
```

```python
usa_grouped.to_csv("clean_data/Death_rate_table.csv", index=False, header=True)
```
