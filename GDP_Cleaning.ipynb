```python
import pandas as pd
from pathlib import Path

gdp_data_to_load = Path("Resources/SASUMMARY__ALL_AREAS_1998_2022.csv")

gdp_data = pd.read_csv(gdp_data_to_load)

gdp_data.head()
```

```python
remove = [str(year) for year in range(1998, 2014)]
data = gdp_data.drop(columns=remove, axis=1)

data.drop(['Region', 'TableName', 'IndustryClassification', 'GeoFIPS', 'LineCode', 'Unit', '2020', '2021', '2022'], axis=1, inplace=True) 

data.head()
```

```python
#line code 1 == gdp
data['Description'] = data['Description'].str.strip()

fil_data = data[data["Description"] == "Real GDP (millions of chained 2012 dollars) 1/"]

#fil_data['GeoName'] = fil_data['GeoName'].str.upper()


fil_data

#make gdp numbers billions of $, currently it is millions
#all uppercase 
```

```python
numeric_columns = ['2014', '2015', '2016', '2017', '2018', '2019']

#fil_data.loc[:, numeric_columns] = fil_data[numeric_columns].apply(pd.to_numeric, errors='coerce', downcast='integer')

fil_data['2014'] = fil_data['2014'].astype(float)
fil_data['2015'] = fil_data['2015'].astype(float)
fil_data['2016'] = fil_data['2016'].astype(float)
fil_data['2017'] = fil_data['2017'].astype(float)
fil_data['2018'] = fil_data['2018'].astype(float)
fil_data['2019'] = fil_data['2019'].astype(float)

fil_data[numeric_columns] = fil_data[numeric_columns] / 1000

#fil_data = fil_data[fil_data["Description"] == "Real GDP (Billions of chained 2012 dollars) 1/"]

fil_data.loc[:, "Description"] = fil_data["Description"].replace("Real GDP (millions of chained 2012 dollars) 1/", "GDP (Billions of Dollars)")

#print(fil_data.dtypes)

fil_data
```

```python
fil_data = fil_data.rename(columns={'GeoName':'state_name', 'Description':'description',  '2014':'amount_2014', '2015':'amount_2015','2016':'amount_2016','2017':'amount_2017','2018':'amount_2018','2019':'amount_2019'})
fil_data
```

```python
fil_data.to_csv('clean_data/GDP_data.csv', index=False)
```
